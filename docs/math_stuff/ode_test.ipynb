# Just testing functionality of `scipy.integrate.solve_ivp`

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def model(t, z):
    x, y, vx, vy = z

    r = np.sqrt(x * x + y * y)

    k = 0.6
    dvx = -k * x / r / r / r
    dvy = -k * y / r / r / r
    dx = vx
    dy = vy
    return [dx, dy, dvx, dvy]

z0 = [0, 10, 0.25, 0]
sol = solve_ivp(model, [0, 400], z0, dense_output=True, vectorized=True, rtol = 1e-8)

z = sol.sol(np.linspace(0, 400, 400))

plt.plot(z.T[:, 0], z.T[:, 1])
plt.xlim([-20, 20])
plt.ylim([-20, 20])
plt.show()
```
